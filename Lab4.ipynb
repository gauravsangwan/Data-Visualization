```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Monthly changes in different food grain stocks.

```python
df = pd.read_csv("/content/drive/MyDrive/Data Visualization/Lab 4/Monthly changes in different food grain stocks.csv")

# Set the index to 'Particulars'
df.set_index('Particulars', inplace=True)

df = df.transpose()
df = df.drop(["Frequency","Unit"])
df
```

```python
# create the plot
fig, ax = plt.subplots(figsize=(8, 5))

# plot the lines
ax.plot(df.index, df['Foodgrains'], color='red', label='Foodgrains')
ax.plot(df.index, df['Foodgrains Rice'], color='green', label='Foodgrains Rice')
ax.plot(df.index, df['Foodgrains Unmilled Paddy in term of Rice'], color='blue', label='Foodgrains Unmilled Paddy in term of Rice')
ax.plot(df.index, df['Foodgrains Wheat'], color='orange', label='Foodgrains Wheat')


# set the x-axis tick labels
ax.set_xticklabels(df.index, rotation=45)

# set the axis labels and legend
ax.set_xlabel('Month')
ax.set_ylabel('Quantity (Ton mn)')
ax.legend(loc='lower right')

# set the title
ax.set_title('Monthly Changes in Food Grain Stocks')

# show the plot
plt.show()
```

#  State wise Colleges of different Ayus Category.

```python
df = pd.read_csv("/content/drive/MyDrive/Data Visualization/Lab 4/State wise Colleges of different Ayus Category.csv")
df["Type of System"].unique()
df =  df[["State or Union Territory","Type of System","No. of Colleges - Govt"]]
df
```

```python
# Set the color palette
cmap = sns.color_palette('rocket_r')

# Create the group bar chart
ax = sns.catplot(data=df, x='State or Union Territory', y='No. of Colleges - Govt', hue='Type of System', kind='bar', palette=cmap, height=6, aspect=2)

ax.set_xticklabels(df['State or Union Territory'], rotation=90)

# Set the chart title and axes labels
plt.title('Number of Government Colleges by State and System Type')
plt.xlabel('State or Union Territory')
plt.ylabel('Number of Colleges - Govt')
# plt.yscale("log")

# Show the chart
plt.show()
```

# Distribution of National, State and District roads.


```python
df = pd.read_csv("/content/drive/MyDrive/Data Visualization/Lab 4/Distribution of National, State and District roads.csv",index_col=0)
df = df[["National Highways","State Highways","District Roads"]]
df
```

```python
# Bipolar sequential palette
colors_bi = sns.color_palette("coolwarm", as_cmap=True)

colors_bi2 = sns.color_palette("GnBu", as_cmap=True)

colors_bi1 = sns.color_palette("Blues", as_cmap = True)

# Create the figure and subplots
fig, axs = plt.subplots(1, 3, figsize=(20, 6))

# National Highways subplot

axs[0] = sns.heatmap(df[['National Highways']], cmap=colors_bi2, center = 5000, ax=axs[0])
axs[0].set_title('National Highways')
axs[0].set_xlabel('')
axs[0].set_ylabel('')
# axs[0].set_xticklabels(axs[0].get_xticklabels(), rotation=45)
# axs[0].set_xscale("log")


# State Highways subplot
sns.heatmap(df[['State Highways']], cmap=colors_bi1, center = 15000, ax=axs[1])
axs[1].set_title('State Highways')
axs[1].set_xlabel('')
axs[1].set_ylabel('')
# axs[1].set_xticklabels(axs[1].get_xticklabels(), rotation=45)

# District Roads subplot
sns.heatmap(df[['District Roads']], cmap=colors_bi, center=30000, ax=axs[2])
axs[2].set_title('District Roads')
axs[2].set_xlabel('')
axs[2].set_ylabel('')
# axs[2].set_xticklabels(axs[2].get_xticklabels(), rotation=45)

sns.set_palette(["#EF476F", "#FFD166", "#06D6A0", "#118AB2"])

# Set the figure title and show the plot
fig.suptitle('Distribution of National, State and District roads', fontsize=18, y=1.1)
plt.show()


```
